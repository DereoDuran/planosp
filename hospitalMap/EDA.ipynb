```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import json
import numpy as np
```

```python
URL_LEITOS_E_INTERNACOES = 'https://raw.githubusercontent.com/seade-R/dados-covid-sp/master/data/plano_sp_leitos_internacoes.csv'
URL_CASOS_E_OBITOS = 'https://raw.githubusercontent.com/seade-R/dados-covid-sp/master/data/dados_covid_sp.csv'
```

```python
class Dashboard():
    
    def __init__(self):
        self.leitos_e_internacoes = pd.read_csv(URL_LEITOS_E_INTERNACOES, sep =';')
        self.casos_e_obitos = pd.read_csv(URL_CASOS_E_OBITOS, sep =';')
```
